```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample DataFrame
data = pd.DataFrame({
    'Num of pregnancies': [1, 2, 3, 4, 5, 2, 1, 3, 2],
    'Biopsy': [0, 1, 0, 1, 0, 1, 0, 1, 1]
})

# Bar plot
sns.catplot(x='Num of pregnancies', y='Biopsy', data=data, kind='bar', height=5, aspect=3)
plt.title('Bar Plot: Number of Pregnancies vs Biopsy')
plt.show()

# Point plot
sns.catplot(x='Num of pregnancies', y='Biopsy', data=data, kind='point', height=5, aspect=3)
plt.title('Point Plot: Number of Pregnancies vs Biopsy')
plt.show()

# Strip plot
sns.catplot(x='Num of pregnancies', y='Biopsy', data=data, kind='strip', height=5, aspect=3)
plt.title('Strip Plot: Number of Pregnancies vs Biopsy')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
cancer_df = pd.read_csv("kag_risk_factors_cervical_cancer.csv")  # Update the path as needed

# Clean the data
cancer_df.replace('?', np.nan, inplace=True)
cancer_df = cancer_df.apply(pd.to_numeric, errors='coerce')
cancer_df.dropna(inplace=True)  # Drop rows with missing values

# Compute the correlation matrix
corr_matrix = cancer_df.corr()

# Plot the heatmap
plt.figure(figsize=(30, 30))  # Adjust the size as needed
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')

# Customize the plot
plt.xticks(rotation=90)  # Rotate x-axis labels
plt.yticks(rotation=0)   # Rotate y-axis labels if needed
plt.title('Heatmap of Correlation Matrix', fontsize=20)
plt.show()
```
